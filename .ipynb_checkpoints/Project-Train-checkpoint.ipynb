```python
import numpy as np
import glob
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn import svm,linear_model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from joblib import dump,load
```

```python
all_data = [] 
cloud_mass_20um = []
rain_mass_20um = []
cloud_number_20um = []
rain_number_20um = []
praut_20um = []
pracw_20um = []
```

```python
all_data
cloud_mass_20um
rain_mass_20um
cloud_number_20um
rain_number_20um
praut_20um
pracw_20um
```

```python
def read_data(filepath):                              
    with open(filepath,'r') as f:
        for a in f:                                      
            temp = a.strip('cloud_mass_20um')           
            temp = temp.strip('rain_mass_20um')
            temp = temp.strip('cloud_number_20um')
            temp = temp.strip('rain_number_20um')
            temp = temp.strip('praut_20um')
            temp = temp.strip('pracw_20um')
            temp = temp.strip()
            temp = temp.strip('[[')
            temp = temp.strip(']]')
            temp = temp.split()                           
            for t in temp:                                
                temp = np.array(t)
                all_data.append(temp)
        data = np.array(all_data).astype(float)                             
        for i in range(len(data)):                                                      
                                                                            
            if i <=3599:
                cloud_mass_20um.append(data[i])
            elif i>=3600 and i<=7199:
                rain_mass_20um.append(data[i])
            elif i>=7200 and i<=10799:
                cloud_number_20um.append(data[i])
            elif i>= 10800 and i<= 14399:
                rain_number_20um.append(data[i])
            elif i>=14400 and i<=17999:
                praut_20um.append(data[i])
            elif i>=18000:
                pracw_20um.append(data[i])
```

```python
#load data
path = glob.glob('./*.txt')
for i in path:
    read_data(i)
    all_data=[]
```

```python
cloud_mass_20um
```

```python
cloud_mass = np.array(cloud_mass_20um)
rain_mass = np.array(rain_mass_20um)
cloud_number = np.array(cloud_number_20um)
rain_number = np.array(rain_number_20um)
praut = np.array(praut_20um)
pracw = np.array(pracw_20um)
```

```python

X = np.stack((cloud_mass,rain_mass,cloud_number,rain_number),axis=0).T
y = np.stack((praut,pracw),axis=0).T
```

```python

print('X shape:',X.shape)
print('y shape:',y.shape)  
```

```python
#data preprocessing
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)

stdscale = StandardScaler()                                   
X_train= stdscale.fit_transform(X_train)                
X_test = stdscale.fit_transform(X_test)                     
y_train_lstm = stdscale.fit_transform(y_train)                   
y_test_lstm = stdscale.fit_transform(y_test)                       

y_train_praut = y_train[:,0].reshape(-1,1)
y_train_pracw = y_train[:,1].reshape(-1,1)
y_test_praut = y_test[:,0].reshape(-1,1)
y_test_pracw = y_test[:,1].reshape(-1,1)

y_train_praut = stdscale.fit_transform(y_train_praut)
y_train_pracw = stdscale.fit_transform(y_train_pracw)
y_test_praut = stdscale.fit_transform(y_test_praut)
y_test_pracw = stdscale.fit_transform(y_test_pracw)
```

```python
y_train_praut
```

```python
X_train
```

```python
#building model
def model_praut(model_name='RF',X_train=X_train, y_train=y_train_praut,
                X_test=X_test, y_test=y_test_praut):
    if model_name == 'SVM':
        clf = svm.SVR(verbose=True)
    elif model_name == 'RF':
        clf = RandomForestRegressor(max_depth=5,random_state=0)
    elif model_name =='LR':
        clf = linear_model.LinearRegression()

    
    y_train_praut_1 = y_train_praut.ravel()                                                         
    y_test_praut_1 = y_test_praut.ravel()
    
    
    scores = cross_val_score(clf,X_train,y_train_praut_1,cv=5)                  
    clf.fit(X_train,y_train_praut_1)                                                                    
    predict_praut = clf.predict(X_test)                                                                 
    y_test_praut1 = stdscale.inverse_transform(y_test_praut_1.reshape(-1,1))
    predict_praut = stdscale.inverse_transform(predict_praut.reshape(-1,1))                         
    r2 = r2_score(y_test_praut1,predict_praut)                                                          
    print('praut_r2: %f' %r2)
    print('praut_5flod_scores:',scores)
        
    dump(clf,'%s_praut.joblib' %model_name)                                                                 
    return predict_praut,y_test_praut1                                                  
```

```python
def model_pracw(model_name='RF', X_train=X_train, y_train=y_train_pracw,
               X_test=X_test, y_test=y_test_pracw):
    if model_name == 'SVM':
        clf_1 = svm.SVR(verbose=True)
    elif model_name == 'RF':
        clf_1 = RandomForestRegressor(max_depth=5,random_state=0)
    elif model_name =='LR':
        clf_1 = linear_model.LinearRegression()
    
    y_train_pracw_1 = y_train_pracw.ravel()
    y_test_pracw_1 = y_test_pracw.ravel()
    scores = cross_val_score(clf_1,X_train,y_train_pracw_1,cv=5)
    clf_1.fit(X_train,y_train_pracw_1)
    predict_pracw = clf_1.predict(X_test)
    y_test_pracw1 = stdscale.inverse_transform(y_test_pracw_1.reshape(-1,1))
    predict_pracw = stdscale.inverse_transform(predict_pracw.reshape(-1,1))
    r2 = r2_score(y_test_pracw1,predict_pracw)
    print('pracw_r2: %f' %r2)
    print('pracw_5flod_scores:', scores)
    
    dump(clf_1,'%s_pracw.joblib' %model_name)
    return predict_pracw,y_test_pracw1
```

```python

predict_praut,true_praut = model_praut(model_name='LR')                               
```
